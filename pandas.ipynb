```python
# series
a = [1,7,3]
myvar = pd.Series(a)
print(myvar)
myvar = pd.Series(a, index = ['x', 'y', 'z'])
print(myvar)
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
```

```python
calories = {"day1": 550, "day2": 440, "day3": 330}
myvar = pd.Series(calories, index = ["day1", "day3"])
print(myvar)
```
