```python
from numpy import random
x = random.randint(100)
print(x)
```

```python
x = random.rand()
print(x)
x = random.randint(100, size = (5))
print(x)
x = random.randint(100, size = (3,5))
print(x)
```

```python
x  = random.choice([1,3,5,7])
print(x)
```

```python
x = random.choice([1,3,5,6], size  = (1,3))
print(x)
```

```python
# data distribution
x = random.choice([1,2,3,4], p=[0.9, 0.1, 0, 0], size = (1,1))
print(x)
```

```python
# random permutation
import numpy as np
arr = np.array([1,3,2,5,4,5])
random.shuffle(arr)
print(arr)
random.permutation(arr)
print(arr)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.kdeplot([0,1,2,3,4,5])
plt.show()
```

```python
# Normal distribution
#loc - (Mean) where the peak of the bell exists.

#scale - (Standard Deviation) how flat the graph distribution should be.

#size - The shape of the returned array.

from numpy import random
x = random.normal(size = (2,3))
print(x)
```

```python
x = random.normal(loc = 1, scale = 2, size = (2,3))
print(x)
```

```python
sns.kdeplot(random.normal(size = 1000))
plt.show()
```

```python
#n - number of trials.

#p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).

#size - The shape of the returned array.
x = random.binomial(n = 10, p=0.5, size = 10)
print(x)
sns.kdeplot(x)
plt.show()
```

```python
sns.histplot(x, kde=False)
plt.show()
```

```python
sns.kdeplot(random.binomial(n=10, p=0.5, size=1000))

plt.show()
```

```python
sns.histplot(random.binomial(n=10, p=0.5, size=1000))

plt.show()
```

```python
sns.kdeplot(random.normal(loc=50, scale=5, size=1000), label='normal')
sns.kdeplot(random.binomial(n=100, p=0.5, size=1000), label='binomial')

plt.show()
```

```python
x = random.poisson(lam =2, size = 10)
print(x)
```

```python
sns.histplot(random.poisson(lam=2, size=1000))

plt.show()
```

# uniform distribution

```python
x = random.uniform(size = (2,3))
print(x)
sns.kdeplot(random.uniform(size =1000))
plt.show()
```

```python
x = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

print(x)
sns.kdeplot(x)
plt.show()
```

```python
sns.kdeplot(random.exponential(size=1000))

plt.show()
```
